# Feature engineering

## Text Preprocessing

### Lowercasing

```python
def lower_text(text):
    return text.lower()
```

```python
text_lower = pd.DataFrame((text.map(lower_text)), columns=["text"])
text_lower
```

### Tokenization - Removing stop words - Punctuation Removal - Stemming/Lemmatization

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
spacy.__version__
```

```python
# text_corrected
```

```python
text_tokenized = []
for text in text_lower["text"]:
    doc = nlp(text)

    txt = ""
    for token in doc:
        if not token.is_stop and not token.is_punct:
            txt += f" {token.lemma_}"
    text_tokenized.append(txt)

text_tokenized = pd.DataFrame(text_tokenized)
text_tokenized
```

## Text Representation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(text_tokenized, nature, test_size=.2, stratify=nature)
```

```python
X_train[0].shape
```

False = 0, True 1

```python
y_train = pd.get_dummies(y_train, drop_first=True, dtype=int)
y_train
```

```python
y_train.value_counts()
```

```python
y_test = pd.get_dummies(y_test, drop_first=True, dtype=int)
y_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()

X_train_cv = v.fit_transform(X_train[0]).toarray()
```

```python
# import joblib

# Save the model
# joblib.dump(v, "vectorizer.jb")
```

# Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
# import joblib

# Save the model
# joblib.dump(model, "naive_baiye_model")
```

```python
X_test_cv = v.transform(X_test[0])
y_pred = model.predict(X_test_cv)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
